## Plot the output as it's being created

````python
import os
import time
from IPython.display import display, clear_output
from PIL import Image

folder = 'StLouisCal310'

while True:
    # Find the newest png file in the folder
    png_files_all = [f for f in os.listdir(folder) if f.endswith('.png')]
    if png_files_all:
        latest_png = max(png_files_all, key=lambda x: os.path.getmtime(os.path.join(folder, x)))
    else:
        latest_png = None
    print(latest_png)
    time.sleep(1)
    if latest_png:
        img_path = os.path.join(folder, latest_png)
        img = Image.open(img_path)
        clear_output(wait=True)
        display(img)
        time.sleep(1)  # Check every second
````
